##### 19. When we make assumptions about how data are distributed, what is the most common assumption?

```python
# Assumption of normality
# If you took a sample, calculated its mean, and wrote this down; then
# took another (independent) sample (from the same population) and got its mean and wrote it
# down; and did this an infinite number of times; then the distribution of the values that you wrote
# down would always be a perfect bell curve. 

# Assumption of independence
# The observations between groups should be independent, 
# which basically means the groups are made up 
# of different people/things/variables. You don’t want one person/things/variables
# appearing twice in two different groups as it could skew your results.
```

##### 27. Generate and interpret 95% confidence intervals for SAT and ACT participation rates.

```python
import scipy.stats as stats
```

```python
stats.ttest_ind(mergedRenamed['sat_participation'], mergedRenamed['act_participation'], axis=0, equal_var=True)
```

```python
# the above cell shows a very high result. It looks like it is approximately
# 4 std dev away from 0.
# the p value is observed to be 0,0002, which less than 0.05. Therefore,
# we can reject the null hypothesis. Thus, identifying that there is a diff
# between SAT and ACT participation rates.
```

```python
stats.t.interval(0.95, len(mergedRenamed['sat_participation'])-1,
              loc=np.mean(mergedRenamed['sat_participation']),
              scale=stats.sem(mergedRenamed['sat_participation']))
```

```python
# By looking at the score above, we can say with 95% confidence that
# the mean of SAT Participation column will fall between 29.8% and 49.7%. 
```

```python
stats.t.interval(0.95, len(mergedRenamed['act_participation'])-1,
              loc=np.mean(mergedRenamed['act_participation']),
              scale=stats.sem(mergedRenamed['act_participation']))
```

```python
# By looking at the score above, we can say with 95% confidence that
# the mean of ACT Participation column will fall between 56.2% and 74.2%. 
```
